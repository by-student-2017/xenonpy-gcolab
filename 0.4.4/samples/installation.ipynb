**Installation**

This tutorial shows the details of installation on Google Golaboratory.

**Step 1**

Install Anaconda and set PATH

```
!wget https://repo.continuum.io/miniconda/Miniconda3-latest-Linux-x86_64.sh
!bash Miniconda3-latest-Linux-x86_64.sh -b -f -p /usr/local
import sys
sys.path.append('/usr/local/lib/python3.6/site-packages/')
```

**step 2**

Install RDKit and check version.

```
!conda install -q -y -c rdkit rdkit=2020.3 python=3.6
import rdkit
print(rdkit.__version__)
```

**step 3**

Install PyTorch and check version.

```
!conda install --yes pytorch torchvision torchaudio cudatoolkit=10.1 -c pytorch
import torch
print(torch.__version__)
```

**step 4**

Install pymatgen and check version. 

```
!conda install --yes --channel conda-forge pymatgen=2020.10.20
import pymatgen
print(pymatgen.__version__)
```

**step 5**

Install XenonPy and check version.

```
!wget https://raw.githubusercontent.com/by-student-2017/xenonpy-gcolab/main/0.4.4/conda_env/xepy36_cuda10.yml
!conda create --yes -n xenonpy python=3.6
!conda env update -n xenonpy -f xepy36_cuda10.yml
!source activate xenonpy
!pip3 install -U xenonpy==0.4.4
import xenonpy
print(xenonpy.__version__)
```
